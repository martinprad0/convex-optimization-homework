```python
import numpy as np
from random import  uniform
from numpy.linalg import norm
from numpy.linalg import inv

"""
PUNTO 3
"""
#FUNCIÓN
def f(x):
    return 100*(x[1]-x[0]**2)**2 + (1-x[0])**2

def f_gradient(x1):
    x = x1
    if len(x) == 1:
        print(x)
    partial1 = -400*(x[0])*(x[1]-x[0]**2) - 2*(1-x[0])
    partial2 = 200*(x[1]-x[0]**2)
    return np.array([partial1,partial2])

def f_hessian(x):
    partial11 = -400*(-3*x[0]**2 + x[1]) + 2
    partial12 = -400*x[0]
    partial21 = -400*x[0]
    partial22 = 200
    return np.matrix([[partial11,partial12],[partial21,partial22]])

#BACKTRACKING
def BT(alpha,c,x_k):
    i = 0
    alpha1 = alpha
    left = f(x_k - alpha1*f_gradient(x_k))
    right = f(x_k) - c*alpha1*f_gradient(x_k)@f_gradient(x_k)
    while left > right:
        ro = uniform(0,1)
        alpha1 = ro*alpha1
        left = f(x_k - alpha1*f_gradient(x_k))
        right = f(x_k) - c*alpha1*f_gradient(x_k)@f_gradient(x_k)
        i += 1
    return alpha1

#A) DESCENSO DEL GRADIENTE 
def GD(x_0,alpha_0,epsilon):
    delta = np.inf
    alpha = alpha_0
    C = 0.5
    x_k = x_0
    i = 0
    while delta > epsilon:
        i+=1
        alpha = BT(alpha,C,x_k)
        x_kplus1 = x_k - alpha*f_gradient(x_k)
        delta = norm(x_kplus1 - np.array([1,1]))
        x_k = x_kplus1
    
    return x_kplus1

#B) NEWTON
def N(x_0,alpha_0,epsilon):
    delta = np.inf
    alpha = alpha_0
    C = 0.5
    x_k = x_0
    i = 0
    while delta > epsilon:
        i+=1
        alpha = BT(alpha,C,x_k)
        x_kplus1 = x_k - alpha*(inv(f_hessian(x_k))@f_gradient(x_k))
        delta = norm(x_kplus1 - np.array([1,1]))
        x_k = np.array(x_kplus1)[0]
    
    return x_kplus1
```

```python
"""
Newton con Backtracking
"""

alpha_0 = 1
x_0 = np.array([1.2,1.2])
epsilon = 0.2

algor = N(x_0, alpha_0,epsilon)
```

```python
"""
PUNTO 5
"""
from random import  uniform
from numpy.linalg import norm

def randomvectors(m,n):
    """
    Genera los vectores aj
    """
    a = np.random.rand(m,n)
    normalvectors = []
    for aj in a:
        k = norm(aj)
        normalvectors.append(1/k*aj)
    return normalvectors

def g(x,c,A):
    t1 = c.dot(x)
    t2 = 0
    for aj in A:
        kj = np.log(round(1 - aj.dot(x),4))
        t2+=kj
    t3 = 0
    for xi in x:
        hj = np.log(1 - xi**2)
        t3+=hj
    return t1 + t2 + t3

def g_gradient(x,c,A,n):
    gradient = []
    for i in range(n): 
        t1 = c[i]
        t2 = 0
        for aj in A:
            kj = aj[i]/(1 - aj.dot(x))
            t2+=kj
        t3 = 2*x[i]/(1 - x[i]**2)
        gradient.append(t1 + t2 + t3)
    return np.array(gradient)

#BACKTRACKING
def BT(alpha,c,x_k1):
    x_k = np.squeeze(x_k1,0)
    left = f(x_k - alpha*f_gradient(x_k))
    right = f(x_k) - c*alpha*f_gradient(x_k)@f_gradient(x_k)
    alpha1 = alpha
    while left > right:
        ro = uniform(0,1)
        alpha1 = ro*alpha1
        left = f(x_k - alpha1*f_gradient(x_k))
        right = f(x_k) - c*alpha1*f_gradient(x_k)@f_gradient(x_k)
    return alpha1


def GD2(x_0,c,A,n,alpha,epsilon):
    delta = np.inf
    x_k = x_0
    i = 0
    while delta > epsilon:
        i+=1
        x_kplus1 = x_k - alpha*g_gradient(x_k,c,A,n)
        #delta = norm(x_kplus1 - np.array([1,1]))
        x_k = x_kplus1
        print(x_kplus1)
    
    return x_kplus1

def GD_backtracking(x_0,alpha_0,epsilon):
    delta = np.inf
    alpha = alpha_0
    C = 0.5
    x_k = x_0
    i = 0
    while delta > epsilon:
        i+=1
        alpha = BT(alpha,C,x_k)
        print("Step ",i,":",alpha)
        x_kplus1 = x_k - alpha*f_gradient(x_k)
        delta = norm(x_kplus1 - np.array([1,1]))
        x_k = x_kplus1
        print(x_kplus1)
        print(delta)
    
    return x_kplus1

```
